```python
import pandas as pd
import matplotlib.pyplot as plt


midwest_df = pd.read_csv('midwest_states.csv')
northeast_df = pd.read_csv('northeastern_states.csv')
south_df = pd.read_csv('southern_states.csv')
west_df = pd.read_csv('western_states.csv')  

avg_midwest_precipitation = midwest_df['Data.Precipitation '].mean()
avg_northeast_precipitation = northeast_df['Data.Precipitation '].mean()
avg_south_precipitation = south_df['Data.Precipitation '].mean()
avg_west_precipitation = west_df['Data.Precipitation '].mean()


regions = ['Midwest', 'Northeast', 'South', 'West']
average_precipitations = [
    avg_midwest_precipitation, 
    avg_northeast_precipitation, 
    avg_south_precipitation, 
    avg_west_precipitation
]


plt.figure(figsize=(10, 6))
plt.bar(regions, average_precipitations, color=['blue', 'green', 'red', 'purple'])
plt.title('Average Data.Precipitation by Region')
plt.xlabel('Region')
plt.ylabel('Average Precipitation')

plt.show()
```
