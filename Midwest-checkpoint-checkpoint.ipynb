```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("usa_weather_clean_data.csv")
df.head()
```

```python
def degree_cal(temp):
    direction=None
    
    num=int(temp)
    if num==0:
        direction="North"
    elif num in [1, 2]:
        direction=" NNE (North-Northeast)"
    elif num in [3, 4]:
        direction=" NE (Northeast)"
    elif num in [5, 6]:
        direction=" ENE (East-Northeast)"
    elif num in [7, 8]:
        direction=" East"
    elif num in [9, 10]:
        direction=" ESE (East-Southeast)"
    elif num in [11, 12]:
        direction=" SE (Southeast)"
    elif num in [13, 14]:
        direction=" SSE (South-Southeast)"
    elif num in [15, 16]:
        direction=" South"
    elif num in [17, 18]:
        direction=" SSW (South-Southwest)"
    elif num in [19, 20]:
        direction=" SW (Southwest)"
    elif num in [21, 22]:
        direction=" WSW (West-Southwest)"
    elif num in [23, 24]:
        direction=" West"
    elif num in [25, 26]:
        direction=" WNW (West-Northwest)"
    elif num in [27, 28]:
        direction=" NW (Northwest)"
    elif num in [29, 30]:
        direction=" NNW (North-Northwest)"
    elif num in [31, 36]:
        direction=" North"
    return direction

```

```python
ans=Midwest_states.groupby(['Date.Month','Station.State'])["Data.Temperature.Avg Temp"].median().reset_index()
```

```python
# Analysing southern states each month which state has average lowest temperature
for month in ans['Date.Month'].unique():  # iterate over unique months
    month_df = ans[ans['Date.Month'] == month]  # filter data for current month
    min_temp = month_df['Data.Temperature.Avg Temp'].min()  # find min temp for current month
    states_with_min_temp = month_df.loc[month_df['Data.Temperature.Avg Temp'] == min_temp, 'Station.State']  # get states with min temp
    for state in states_with_min_temp:
        print(f"Month: {month}, State: {state}, Temperature: {min_temp}")
```
