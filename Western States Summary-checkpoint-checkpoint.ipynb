```python
def degree_cal(temp):
    direction=None
    
    num=int(temp)
    if num==0:
        direction="North"
    elif num in [1, 2]:
        direction=" NNE (North-Northeast)"
    elif num in [3, 4]:
        direction=" NE (Northeast)"
    elif num in [5, 6]:
        direction=" ENE (East-Northeast)"
    elif num in [7, 8]:
        direction=" East"
    elif num in [9, 10]:
        direction=" ESE (East-Southeast)"
    elif num in [11, 12]:
        direction=" SE (Southeast)"
    elif num in [13, 14]:
        direction=" SSE (South-Southeast)"
    elif num in [15, 16]:
        direction=" South"
    elif num in [17, 18]:
        direction=" SSW (South-Southwest)"
    elif num in [19, 20]:
        direction=" SW (Southwest)"
    elif num in [21, 22]:
        direction=" WSW (West-Southwest)"
    elif num in [23, 24]:
        direction=" West"
    elif num in [25, 26]:
        direction=" WNW (West-Northwest)"
    elif num in [27, 28]:
        direction=" NW (Northwest)"
    elif num in [29, 30]:
        direction=" NNW (North-Northwest)"
    elif num in [31, 36]:
        direction=" North"
    return direction

```

```python
df_wide = df.pivot_table(index="Date.Month", columns="Station.State", values="Data.Temperature.Avg Temp")
```

```python
wind_speed_wide = df.pivot_table(index="Date.Month", columns="Station.State", values="Data.Wind.Speed")
```

```python
wind_speed_wide
```

```python
plt.figure(figsize=(10,5))
ax=sns.lineplot(data=wind_speed_wide)
sns.move_legend(ax,"center right",bbox_to_anchor=(1.2, 0.5)) 
plt.title("Wind Speed IN Western STATES",fontsize=14)
plt.axhline(5.02,color="Red",linestyle="--")
plt.text(11,5.7,"Average",
        fontsize=13,
        color="Red",
        )
plt.xlabel("Months")
plt.ylabel("Windspeed in mph")
plt.savefig("avg_wind west (line)")
plt.show()
```

- Most of the states wind speed is around the average (red dashed line)
- Hawaii has high average wind speed , in summer seasons more winds due to  Trade winds are formed due to the Earth's rotation.   This rotation causes the air from the northern hemisphere to slant towards the equator in a southwesterly direction.
- Idaho and Washingtion are mostly windy during winters
- Alaska is windy at starting of the year
- Colorado is in Tornado valley hence the  winds around the average 
- we can observe a seasonal trend where the wind speed for most of the states  other than Hawaii , New Mexico , Wyoming are lower or around the average throughout the year .
