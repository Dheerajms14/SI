```python
def degree_cal(temp):
    direction=None
    
    num=int(temp)
    if num==0:
        direction="North"
    elif num in [1, 2]:
        direction=" NNE (North-Northeast)"
    elif num in [3, 4]:
        direction=" NE (Northeast)"
    elif num in [5, 6]:
        direction=" ENE (East-Northeast)"
    elif num in [7, 8]:
        direction=" East"
    elif num in [9, 10]:
        direction=" ESE (East-Southeast)"
    elif num in [11, 12]:
        direction=" SE (Southeast)"
    elif num in [13, 14]:
        direction=" SSE (South-Southeast)"
    elif num in [15, 16]:
        direction=" South"
    elif num in [17, 18]:
        direction=" SSW (South-Southwest)"
    elif num in [19, 20]:
        direction=" SW (Southwest)"
    elif num in [21, 22]:
        direction=" WSW (West-Southwest)"
    elif num in [23, 24]:
        direction=" West"
    elif num in [25, 26]:
        direction=" WNW (West-Northwest)"
    elif num in [27, 28]:
        direction=" NW (Northwest)"
    elif num in [29, 30]:
        direction=" NNW (North-Northwest)"
    elif num in [31, 36]:
        direction=" North"
    return direction

```

# Temperature

```python
df_wide = df.pivot_table(index="Date.Month", columns="Station.State", values="Data.Temperature.Avg Temp")
```

```python
df_wide
```

```python
#min temp
temp_min=df.pivot_table(index="Date.Month", columns="Station.State", values="Data.Temperature.Min Temp")
```

```python
temp_min
```

```python
plt.figure(figsize=(10,5))
ax=sns.lineplot(data=temp_min,lw=2)
sns.move_legend(ax,"center right",bbox_to_anchor=(1.2, 0.5))
plt.title("AVERAGE TEMPERATURE (North Eastern)",fontsize=14)
plt.xlabel("Months")
plt.ylabel("temperature in F")
plt.axhline(53,color="Red",lw=3,linestyle="--")
plt.text(11,55,"Average",
        fontsize=12,
        color="Red")
plt.show()
```

- temperature rise during the summer 
- the North Eastern states are usually experience warm summers and cold winters
- Rhode Island,Connecticut,Pennysylvania have high avg temperatures during the summer season
- the states are closer to the east coast hence it is colder durin winters and warm during the summers
