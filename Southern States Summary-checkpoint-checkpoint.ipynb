# rainfall

```python
rain_wide = df.pivot_table(index="Date.Month", columns="Station.State", values="Data.Precipitation ")
```

```python
rain_wide
```

```python
plt.figure(figsize=(10,5))
ax=sns.lineplot(data=rain_wide)
sns.move_legend(ax,"center right",bbox_to_anchor=(1.2, 0.5))
plt.title("Rainfall IN SOUTHERN STATES")
plt.xlabel("Months")
plt.ylabel("Rainfall inches")
plt.savefig("avg-rainfall (line)-South")
plt.show()
```

- Louisiana and Mississippi have high average rainfall
- In Louisiana the rainfall was in peak during August which resulted in  floods.
- Links <a href="https://www.climate.gov/news-features/event-tracker/august-2016-extreme-rain-and-floods-along-gulf-coast">Noaa report </a> <br> <a href="https://www.climate.gov/news-features/event-tracker/maya-express-behind-gulf-coast-soaking">Maya express </a>

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=rain_wide["Louisiana"],color="Red")
sns.lineplot(data=rain_wide["Mississippi"],color="Green")
sns.lineplot(data=rain_wide["Arkansas"],color="Orange")
sns.lineplot(data=rain_wide["Texas"],color="Blue")
plt.title("States affected with floods")
plt.ylabel("Rainfall")
plt.xlabel("Months")
plt.savefig("floods - South")
plt.show()
```

- we can observe the peak in august which caused floods 
- Heavy rainfall occurred across Louisiana, Texas, Arkansas, and Mississippi in March 2016 as a result of a slow-moving southward dip in the jetstream, funneling tropical moisture into parts of the Gulf Coast States and the Mississippi River Valley. The storm caused major flooding in the northwestern and southeastern parts of Louisiana and in eastern Texas.

```python
import pandas as pd


data = {'state': ['Alabama', 'Arkansas', 'Delaware', 'Florida', 'Georgia', 'Kentucky', 'Louisiana', 'Maryland', 'Mississippi', 'North Carolina', 'Oklahoma', 'South Carolina', 'Tennessee', 'Texas', 'Virginia', 'West Virginia'],
        'avg_wind_speed': [4.82, 3.3, 6.76, 6.3, 4.67, 4.97, 5.1, 5.64, 4.74, 5.25, 8.48, 4.98, 4.06, 7.71, 5.9, 4.14]}

# Create DataFrame
wind_df = pd.DataFrame(data)


```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=wind_df["state"],y=wind_df["avg_wind_speed"],data=wind_df)
plt.xticks(rotation=45)
plt.title("Average wind speed state wise in Southern states",fontsize=20)
plt.axhline(5.8,lw=3,color='red',linestyle="--")
plt.text(11,6.2,"average",
        fontsize=14,
        color="red")
plt.ylabel("average wind speed in mph")
plt.xlabel("State")
plt.savefig("Avg_wind (Southern)")
plt.show()
```
